#Statistics Advance Part 1 Answers

###1. What is a random variable in probability theory?

-> A random variable is a function that assigns numerical values to the outcomes of a random experiment. It quantifies outcomes in a sample space and can be either discrete (like number of heads) or continuous (like time or temperature).

###2. What are the types of random variables?

-> There are two types:

- Discrete random variable: Takes countable values (e.g., number of students).

- Continuous random variable: Takes infinite values within a range (e.g., height, weight).

###3. What is the difference between discrete and continuous distributions?

-> Discrete distributions represent probabilities at specific values while Continuous distributions describe the probability of outcomes over an interval and are represented by a probability density function.

###4. What are probability distribution functions (PDF)?
-> A PDF describes how the probabilities of a continuous random variable are distributed. It gives the likelihood that the variable falls within a particular range. The area under the PDF curve between two points gives the probability of falling within that range.

###6. What is a discrete uniform distribution?
-> It’s a distribution where all possible outcomes are equally likely. For example, in rolling a fair die, each outcome (1-6) has a probability of 1/6.

###7. What are the key properties of a Bernoulli distribution?

- Only two outcomes: success (1) and failure (0)

- Probability of success = p, failure = 1−p

- Mean = p, Variance = p(1−p)
It models a single trial like flipping a coin.

###8. What is the binomial distribution, and how is it used in probability?
-> It gives the probability of having k successes in n independent Bernoulli trials. It's used in quality control, surveys, and experiments where outcomes are binary.

###9. What is the Poisson distribution and where is it applied?
-> Models the number of events in a fixed interval of time or space, assuming they occur independently at a constant average rate. Used in scenarios like call arrivals, decay rates, or web traffic.

###10. What is a continuous uniform distribution?
A distribution where all outcomes in a continuous range [a, b] are equally likely. The PDF is constant:

$f(x)=1/(b-a)$ for $a \leq x$ $\leq $b


###11. What are the characteristics of a normal distribution?

- Bell-shaped and symmetric

- Mean = Median = Mode

- Defined by mean (μ) and standard deviation (σ)

- 68-95-99.7 rule applies for standard deviations from the mean

###12. What is the standard normal distribution, and why is it important?
-> It is a normal distribution with mean 0 and standard deviation 1.

It's used for standardization and calculating Z-scores to compare different datasets.

###13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
-> CLT states that the distribution of sample means approaches a normal distribution as the sample size becomes large, regardless of the population's distribution. This is essential for making inferences about population parameters.

###14. How does the Central Limit Theorem relate to the normal distribution?
-> It explains why the sampling distribution of the sample mean becomes approximately normal for large samples, enabling the use of normal-based methods in hypothesis testing and confidence intervals.

###15. What is the application of Z statistics in hypothesis testing?
-> Z statistics are used to test hypotheses about population means when the population standard deviation is known. It helps assess whether a sample mean significantly differs from a population mean.

###16. How do you calculate a Z-score, and what does it represent?

Z- score is calculated as:

It represents how many standard deviations a data point (X) is from the population mean (μ).

###17. What are point estimates and interval estimates in statistics?

-> Point estimate: A single value estimate of a parameter (e.g., sample mean).

-> Interval estimate: A range (confidence interval) within which the parameter lies with a specified probability.

###18. What is the significance of confidence intervals in statistical analysis?
-> They provide a range of values likely to contain the population parameter. A 95% confidence interval means that if we repeat the process 100 times, the true parameter would lie within the interval 95 times.

###19. What is the relationship between a Z-score and a confidence interval?
-> The Z-score sets the width of the confidence interval by determining how much uncertainty is allowed. Higher confidence means a larger Z-score and wider interval, ensuring a higher chance of capturing the true parameter.

Z-scores determine the margin of error used in confidence intervals. For example, for a 95% confidence interval, Z = 1.96 is used in the formula.

###20. How are Z-scores used to compare different distributions?
-> They standardize values from different distributions, enabling direct comparison by removing units and expressing values in terms of standard deviations.

###21. What are the assumptions for applying the Central Limit Theorem?

- Random sampling

- Independent observations

- Finite population variance

- Sample size sufficiently large (typically n > 30)

###23. How does a probability distribution relate to the expected outcome of a random variable?
-> The probability distribution defines the likelihood of each outcome, and the expected value is the weighted average of all outcomes based on their probabilities.
